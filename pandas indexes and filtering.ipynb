## Using function to do advance searching of how many people know Python in each country

```python
country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```

```python
country_respondents = df['Country'].value_counts()
country_respondents
```

```python
country_uses_python = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_uses_python
```

## We can use ' concat ' function to combine two or more series

## we use axis = columns because by deafult it is set to rows

```python
python_df = pd.concat([country_respondents,country_uses_python],axis = 'columns',sort = False)
python_df
```

```python
python_df.rename(columns={'Country':"NumRespondents",'LanguageWorkedWith':"NumkonwsPython"},inplace = True)
```

```python
python_df
```

## To get percentage of people in each country who knows python

```python
python_df['PercentageknowsPython'] = (python_df['NumkonwsPython']/python_df['NumRespondents']) * 100
python_df
```

```python
python_df.sort_values(by='PercentageknowsPython',ascending=False,inplace=True)
```
